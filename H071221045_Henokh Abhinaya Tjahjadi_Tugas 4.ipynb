```python


print("=========================================================")
print("                       Kalkulator                        ")
print("=========================================================")
print("Pilih Opsi:                                              ")
print("1]. [ Tambah                                            ]")
print("2]. [ Kurang                                            ]")
print("3]. [ Kali                                              ]")
print("4]. [ Bagi                                              ]")
print("5]. [ Akar                                              ]")
print("6]. [ Pangkat                                           ]")
print("---------------------------------------------------------")

import math

def tambah(x, y):
    return x + y

def kurang(x, y):
    return x - y

def kali(x, y):
    return x * y

def bagi(x, y):
    return x / y

def akar(x):
    return math.sqrt(x)

def pangkat(x, y):
    return x ** y

while True:
    opsi = input(">Pilih Opsi: ")

    if opsi in ('1', '2', '3', '4', '5', '6'):
        print("---------------------------------------------------------")
        if opsi == '1':
            num1 = float(input("Input angka pertama: "))
            num2 = float(input("Input angka kedua  : "))
            print("Angka pertama adalah ", num1)
            print("Angka kedua adalah ", num2)
            print("Hasil Dari", num1, "+", num2, "adalah", tambah(num1, num2))

        elif opsi == '2':
            num1 = float(input("Input angka pertama: "))
            num2 = float(input("Input angka kedua  : "))
            print("Angka pertama adalah ", num1)
            print("Angka kedua adalah ", num2)
            print("Hasil Dari", num1, "-", num2, "adalah", kurang(num1, num2))

        elif opsi == '3':
            num1 = float(input("Input angka pertama: "))
            num2 = float(input("Input angka kedua  : "))
            print("Angka pertama adalah ", num1)
            print("Angka kedua adalah ", num2)
            print("Hasil Dari", num1, "*", num2, "adalah", kali(num1, num2))

        elif opsi == '4':
            num1 = float(input("Input angka pertama: "))
            num2 = float(input("Input angka kedua  : "))
            print("Angka pertama adalah ", num1)
            print("Angka kedua adalah ", num2)
            print("Hasil Dari", num1, "/", num2, "adalah", bagi(num1, num2))
        
        elif opsi == '5':
            num1 = float(input("Input angka yang ingin diakarkan: "))
            print("Angkanya adalah ", num1)
            print("Hasil Dari", "âˆš", num1, "adalah", akar(num1))

        elif opsi == '6':
            num1 = float(input("Input angka  : "))
            num2 = float(input("Input Pangkat: "))
            print("Angka pertama adalah ", num1)
            print("Angka kedua adalah ", num2)
            print("Hasil Dari", num1, "^", num2, "adalah", pangkat(num1, num2))

        lanjut = input("Apakah ingin masih menggunakan kalkulator? (y/n): ")
        if lanjut == "n":
            print("=========================================================")
            print("        Terima Kasih Telah Menggunakan Kalkulator        ")
            print("=========================================================")
            break

    
    else:
        print("Invalid Input")
```
